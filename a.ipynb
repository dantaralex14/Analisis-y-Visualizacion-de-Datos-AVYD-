```python
# Filtrar los datos según los criterios especificados
filtro_accidentes = (df['indicador'] == 'Accidentes de tránsito terrestre en zonas urbanas y suburbanas') & (df['unidad_medida'] == 'Accidentes')
filtro_heridas = (df['indicador'] == 'Víctimas heridas en los accidentes de tránsito') & (df['unidad_medida'] == 'Víctimas')
filtro_muertas = (df['indicador'] == 'Víctimas muertas en los accidentes de tránsito') & (df['unidad_medida'] == 'Víctimas')

# Aplicar los filtros y sumar las cantidades por fecha
df_suma = df[filtro_accidentes | filtro_heridas | filtro_muertas].groupby(['año', 'indicador', 'unidad_medida']).agg({'valor': 'sum'}).reset_index()
```

```python
import matplotlib.pyplot as plt

# Crear una figura y ejes para los subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 8))

# Iterar sobre los diferentes indicadores y graficar cada uno
for i, (indicador, datos) in enumerate(df_suma.groupby('indicador')):
    axs[i].plot(datos['año'], datos['valor'], label=indicador)
    axs[i].set_xlabel('Año')
    axs[i].set_ylabel('Cantidad')
    axs[i].set_title(f'{indicador} por Año')
    axs[i].grid(True)
    axs[i].legend()

# Ajustar el diseño de los subplots
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import numpy as np

# Calcula la media y la desviación estándar para cada indicador
mean_std_per_indicador = df_suma.groupby('indicador')['valor'].agg(['mean', 'std'])

# Define un umbral para la detección de anomalías (por ejemplo, 3 desviaciones estándar)
umbral = 3

# Detecta las anomalías para cada indicador
anomalias_por_indicador = {}
for indicador, datos in df_suma.groupby('indicador'):
    mean = mean_std_per_indicador.loc[indicador, 'mean']
    std = mean_std_per_indicador.loc[indicador, 'std']
    lower_bound = mean - umbral * std
    upper_bound = mean + umbral * std
    anomalias = datos[(datos['valor'] < lower_bound) | (datos['valor'] > upper_bound)]
    anomalias_por_indicador[indicador] = anomalias

# Muestra las anomalías detectadas
for indicador, anomalias in anomalias_por_indicador.items():
    print(f"Anomalías para {indicador}:")
    print(anomalias)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Supongamos que df_suma es tu DataFrame con los datos resumidos por año e indicador

# Crear DataFrame para las gráficas
df_graficas = df_suma.pivot(index='año', columns='indicador', values='valor').reset_index()

# Utilizar Isolation Forest para detectar anomalías en cada indicador
plt.figure(figsize=(15, 18))

for i, indicador in enumerate(df_graficas.columns[1:]):
    # Crear subplot para cada indicador
    plt.subplot(len(df_graficas.columns)-1, 1, i+1)

    # Extraer valores para el indicador actual
    valores = df_graficas[indicador].values.reshape(-1, 1)

    # Detectar anomalías con Isolation Forest
    iso_forest = IsolationForest(contamination=0.05) # Suponemos que aproximadamente el 2% de los datos son anomalías
    anomalies = iso_forest.fit_predict(valores)

    # Marcar anomalías en la gráfica
    plt.plot(df_graficas['año'], valores, label=indicador)
    plt.scatter(df_graficas.loc[anomalies == -1, 'año'], valores[anomalies == -1], color='red', label='Anomalía', marker='x', s=100)

    # Añadir etiquetas y título
    plt.xlabel('Año')
    plt.ylabel('Cantidad')
    plt.title(f'{indicador} con Anomalía Detectada')
    plt.legend()
    plt.grid(True)

# Ajustar espaciado entre subgráficas
plt.tight_layout()

# Mostrar las gráficas
plt.show()
```

```python
df_cantidad_accidentes = df_suma[df_suma['indicador'] == 'Accidentes de tránsito terrestre en zonas urbanas y suburbanas'][['año','valor']]

# Convertir la columna 'valor' en una serie temporal
df_cantidad_accidentes['año'] = pd.to_datetime(df_cantidad_accidentes['año'], format='%Y')

# Establecer 'año' como el índice del DataFrame
df_cantidad_accidentes.set_index('año', inplace=True)

# Asegurar la frecuencia de muestreo
df_cantidad_accidentes = df_cantidad_accidentes.resample('A').sum()

# Aplicar la descomposición estacional al DataFrame de accidentes
from statsmodels.tsa.seasonal import seasonal_decompose

descomposicion_accidentes = seasonal_decompose(df_cantidad_accidentes['valor'], model='additive', period=7)
```

```python
df_cantidad_accidentes.shape
```

```python
# Visualización de las componentes de la descomposición estacional
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
# Se traza la serie temporal original en el primer subplot
descomposicion_accidentes.observed.plot(label='Original')
plt.title('Original - Cantidad de accidentes')
plt.legend()

plt.subplot(4, 1, 2)
# Se traza la componente de tendencia en el segundo subplot
descomposicion_accidentes.trend.plot(label='Tendencia')
plt.title('Tendencia - Cantidad de accidentes')
plt.legend()

plt.subplot(4, 1, 3)
# Se traza la componente estacional en el tercer subplot
descomposicion_accidentes.seasonal.plot(label='Estacionalidad')
plt.title('Estacionalidad - Cantidad de accidentes')
plt.legend()

plt.subplot(4, 1, 4)
# Se trazan los residuos en forma de dispersión en el cuarto subplot
plt.scatter(descomposicion_accidentes.resid.index, descomposicion_accidentes.resid, label='Residual')
plt.axhline(0, color='black')  # Agrega una línea horizontal en el valor cero
plt.title('Residual - Cantidad de accidentes')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

suavizado_simple = df_cantidad_accidentes['valor'].rolling(window=12).mean()

# Primera diferenciación
diferenciacion = df_cantidad_accidentes['valor'].diff().dropna()

# Visualización de las componentes de suavizado simple y primera diferenciación
plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
df_cantidad_accidentes['valor'].plot(label='Original')
suavizado_simple.plot(label='Suavizado Simple', color='red')
plt.title('Suavizado Simple vs Original')
plt.legend()

plt.subplot(2, 1, 2)
diferenciacion.plot(label='Primera Diferenciación', color='green')
plt.title('Primera Diferenciación')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Prueba de Dickey-Fuller para la serie original
resultado_original = adfuller(df_cantidad_accidentes['valor'])
if resultado_original[1] <= 0.05:
    print('La serie original es estacionaria')
else:
    print('La serie original no es estacionaria')

# Prueba de Dickey-Fuller para la serie diferenciada
resultado_diferenciada = adfuller(diferenciacion)
if resultado_diferenciada[1] <= 0.05:
    print('La serie diferenciada es estacionaria')
else:
    print('La serie diferenciada no es estacionaria')

```

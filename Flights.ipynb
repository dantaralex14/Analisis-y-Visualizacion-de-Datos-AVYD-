## Dataset Flights ##

```python
# Importamos librerias
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Cargamos datasets
flights = sns.load_dataset("flights")
```

```python
# analizamos dataset
flights.head()
```

```python
flights.shape
```

```python
# Le cinq dernieres register
flights.tail()
```

```python
# cada fila un mes y cada columna y tendremos la evolucion de los vuelos
# convertimos a un formato ancho donde cada celda es el numero de pasajeros
# en indice que los meses y en la columnas cada de los años
flights_wide = flights.pivot_table(values="passengers", index = 'month', columns= 'year')
# revisar conversion
flights_wide.head()
```

## Mapa de calor ##

Un mapa de calor es una representacion grafica de datos donde los valores de una matriz se representan como colores. Se utilizan comunmente para visualizar la intensidad de los datos en una matriz de datos, son utiles para mostrar la distribucion y relacion entre dos variables

```python
# crear mapa de calor
sns.heatmap(flights_wide, annot=True)
```

```python
# Crear mapa de calor
# Texto en notacion cientifica convertimos a entero
ax = sns.heatmap(flights_wide, annot=True, fmt="g")
ax.set_title('Evolución de Vuelos por Mes/Año', size=14)
# Cambiar tamaño de las leyendas del eje x y eje y
ax.set_xticklabels(ax.get_xticklabels(), fontsize=6, rotation=0)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=6)
```

```python
# Sumar el total de los vuelos por año
flights_total = flights_wide.sum().index
# Crear un gráfico de barras con la suma de pasajeros por año con seaborn
ax = sns.barplot(x=flights_total, y=flights_wide.sum().values, hue=flights_total, palette='summer_r', legend=False)
# Cambiar el el título del gráfico
ax.set_title('Total Passengers per Year')
# Cambiar el tamaño del gráfico
ax.figure.set_size_inches(10, 5)
```

```python
import plotly.express as px
import pandas as pd
```

```python
wind_df = px.data.wind()
```

```python
print(wind_df.head())
print(wind_df.info())
```

```python
climate_vars = ['direction', 'frequency', 'strength']
```

```python
scatter_fig = px.scatter_polar(wind_df, r="frequency", theta="direction",
                               color="strength", title="Wind Frequency and Strength",
                               labels={"frequency": "Frequency", "direction": "Direction",
                                       "strength": "Strength"})
scatter_fig.write_html("wind_scatter.html")
```

```python
line_fig = px.line_polar(wind_df, r="frequency", theta="direction",
                         title="Wind Frequency Trend by Direction",
                         labels={"frequency": "Frequency", "direction": "Direction"},
                         color="strength", line_close=True)
line_fig.write_html("wind_line.html")
```

```python
bar_fig = px.bar_polar(wind_df, r="frequency", theta="direction",
                       title="Wind Frequency Distribution by Direction",
                       labels={"frequency": "Frequency", "direction": "Direction"},
                       color="strength")
bar_fig.write_html("wind_bar.html")
```

#### **Seaborn** ####

Seaborn es una de las bibliotecas de visualizacion mas importante de python

```python
# Importamos seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
# Graficamos
sns.relplot(x=x, y = x**2, kind='scatter')
```

```python
sns.lineplot(x=x, y=x**2)
```

#### **Plot Styles** ####

```python
contexts = ['notebook', 'paper', 'talk', 'poster']
# Ejemplo de contexto tipo notebook (usado por defecto)
sns.set_context(contexts[0])
x = np.linspace(0,10,100)
sns.relplot(x=x, y = x**2, kind='scatter')
```

```python
# Parametros del contexto
# Propiedades de las graficas que cambian los contextos
sns.plotting_context()
```

```python
# Seaborn dispone de una biblioteca de paleta 
# Modificar los colores mediante una paleta de colores
# Paleta de colores que usamos por default
sns.color_palette()
```

```python
sns.set_palette('inferno')
sns.color_palette()
```

```python
# Vemos los cambios de paleta aplicados a una gráfica inferno
sns.set_palette('magma', 15)
sns.set_context(contexts[0]) # Default notebook
x  = np.linspace(0,10,100)
sns.relplot(x=x, y=x**2, kind='scatter')
# Como solo tenemos una sola gráfica ha escogido un solo color
# del estilo inferno para pintarla
```

## Estilos de Gráficos de seaborn ##

```python
styles = ['White','dark','whitegrid','darkgrid']
```

```python
# Prueba con cada uno de ellos
sns.set_style(styles[1])
# Probemos los distintos estilos
x = np.linspace(0, 10, 100)
sns.relplot(x=x, y= x**2, kind='scatter')
```

```python
# Podemos cambiar los tres parámetros directamente usando un theme
# Contexto, styles y  paleta
sns.set_theme(context='notebook', style='darkgrid', palette='inferno', font='sans-serif' )
x = np.linspace(0, 10, 100)
sns.relplot(x=x, y= x**2, kind='scatter')
```

## Cargar dataset Titanic ##

```python
# Titanic Dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Who was on the titanic? ##

```python
# Vamos a crear cuatro diagramas de barras
fig, axes = plt.subplots(2,2, figsize=(12,8))
# Para el campo edad
sns.histplot(data=titanic, x='age', ax=axes[0,0])
# Para el grupo demográfico
sns.histplot(data=titanic, x='who', ax=axes[0,1], color='gold')
# Para la clase en la que viajan los pasajeros
sns.histplot(data=titanic, x='class', ax=axes[1,0], color='skyblue')
# Para conoce los puntos de embarque de los pasajeros
sns.histplot(data=titanic, x='embark_town', ax=axes[1,1], color='teal')
# Ponemos un titulo a la gráfica
fig.suptitle('Passengers Demographics', size= 24)
# Cambiemos el color de las gráficas
```

```python
sns.kdeplot(data=titanic, x='age')
```

```python
# Descomponiendo la distribucion segun la clase en la que viajan
sns.kdeplot(data=titanic, x='age',hue='class')
```

```python
ax = sns.kdeplot(data=titanic,x='age',hue='class',shade=True,palette='inferno')
#Agregamos un titulo
ax.set_title('Pasenger Age By Class', size=20)
```

```python
# Distribución Acumulada de la Edad por Clase
ax = sns.ecdfplot(data=titanic, x='age', hue='class')
ax.set_title('Cumulative Age Distribution By Class', size=18)
```

```python
sns.displot(data=titanic, x='who', col='class', hue='alone', multiple='stack', palette='inferno')
```

```python
sns.histplot(data=titanic, x='fare', bins=30)
```

```python
sns.histplot(data=titanic, x='fare', hue='class', multiple='stack', bins=30)
```

```python
# Creamos una grafica de caja para visualizar los sobrevivientes del titanic
sns.catplot(data=titanic, x='alive', y='fare', col='class', kind='box')
```

```python
# Vamos a modificar la escala mediante una función logaritmica para quitar lo comprimido a las gráficas
face_grid = sns.catplot(data=titanic.query('fare >0'), x='alive',  y='fare', col='class', kind='box', aspect=0.5, palette='magma')
face_grid.set(yscale='log')
```

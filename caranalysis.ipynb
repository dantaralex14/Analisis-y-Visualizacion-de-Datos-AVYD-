```python
encabezados = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration",
              "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base",
              "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size",
              "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm",
              "city-mpg", "highway-mpg", "price"]

df.columns = encabezados
```

```python
# Reemplaza los valores de las celdas "?" con el valor de "Nan"
df.replace("?",np.NaN)
```

```python
new_headers = {
    "symboling", "normalized-losses", "make", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base",
    "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size",
    "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm",
    "city-mpg", "highway-mpg", "price"
}

df = df.rename[columns=new_headers]
```

```python
df["city-kpl"] = ["highway-kpl"] *0.46
df
```

```python
#Obtener estadisticas por columna
#La media es el promedio de todo el conjunto de datos
df["city-mpg"].mean()
```

```python
#La moda es el valor que aparece con mayor frecuencia en un conjunto de datos
#si no hay un valor que aparezca mas veces que los demas, entonces no hay moda
df["city-mpg"].mode()
```

```python
# La desviacion estandar es una medida de dispersion que 
# Primero calculamos la media de los datos
# Luego, calculamos la diferencia entre cada valor y la media,
# Sumamos todas estas diferencias al cuadrado
# Dividimos esta suma por el numero total de valores
# Finalmente tomamos la raiz cuadrada del resultado
df["city-mpg"].std()
```

```python
# importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt

# Generar 5 datos aleatorios
datos = np.random.randint(1, 20, size=5) # Genera 5 numeros enteros aleatorios

# Crear una lista de etiquetas para las barras
etiquetas = ['Auto 1', 'Auto 2', 'Auto 3', 'Auto 4', 'Auto 5']

# Crear la grafica de barras
plt.bar(etiquetas, datos, color='blue')

# Agregar titulo y etiquetas a los ejes
plt.title('Ejemplo de Gráfica de Barras')
plt.xlabel('Automóviles')
plt.ylabel('Valores')

# Mostrar la grafica
plt.grid(True, linestyle='--', alpha=0.7) # Agregar una cuadricula
plt.tight_layout() # Ajustar diseño
plt.show()

```

```python
# Gráfica de distribución de marca de automóviles
# Obtener el número de autos  de cada marca
nummarcas = df["make"].value_counts()

# Graficar la distribución de marcas de automóviles en un gráfico de barras
ax = nummarcas.plot(kind="bar", figsize=(10, 6), color='skyblue', edgecolor='black')
nummarcas.plot(kind="bar", figsize=(10, 6), color='skyblue', edgecolor='black')
# Agregar título y etiquetas a los ejes
plt.title("Distribución de Marcas de Automóviles")
plt.xlabel("Marcas")
plt.ylabel("Número")
# Agregar la cantidad por marca arriba de cada barra
for i, v in enumerate(nummarcas):
   ax.text(i, v + 0.2, str(v), ha='center', va='bottom')

# Mostrar la gráfica
# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45)
# Agregar una cuadrícula horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Ajustar diseño
plt.tight_layout()  
plt.show()
```

```python
#RENDIMIENTO DE AUTOMOVILES
# Crear una gráfica de rendimiento de cada marca en base al kilometraje
plt.figure(figsize=(12, 8))  # Aumentar el tamaño de la figura

# Agrupar los datos por marca y calcular el kilometraje promedio en ciudad y autopista
avg_mileage = df.groupby('make')[['city-mpg', 'highway-mpg']].mean()

# Graficar el rendimiento de las marcas de automóviles
avg_mileage.plot(kind='bar', width=0.8)  # Ajustar el ancho de las barras
plt.title('Rendimiento de Marcas de Automóviles')
plt.xlabel('Marca')
plt.ylabel('Kilometraje Promedio (mpg)')
plt.xticks(rotation=45, ha='right')  # Rotar y ajustar la alineación de las etiquetas x
plt.legend(['Ciudad', 'Autopista'])
plt.tight_layout()
plt.show()
```
